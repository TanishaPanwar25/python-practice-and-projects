```python
# Select required columns
survey_df = survey_df[["Gender", "Age"]]

survey_df = survey_df[~survey_df["Gender"].str.contains(";", na=False)]

# Create age groups
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ["<18", "18-25", "26-35", "36-45", "46-55", "56-65", "65+"]

survey_df["AgeGroup"] = pd.cut(survey_df["Age"], bins=age_bins, labels=age_labels)

# Gender + Age group analysis
result = (
    survey_df
    .groupby(["Gender", "AgeGroup"])
    .size()
    .reset_index(name="Count")
)

print(result)
```
