```python
#Exception by zerodivitionErrror
try:
    divideby=int(input("With what value you want to divide number/10"))
    result=10/divideby
    print("result",result)
except ZeroDivisionError as e:
    print("some wnet wrong",e)
```

```python
#Multiple exception/else/finally
try:
    divideby=input("With what value you want to divide number/10")
    result=10/int(divideby)
except ZeroDivisionError as e:
    print("some wnet wrong",e)
    print("some wnet wrong",e.args)
except TypeError as e:
    print("some wnet wrong",e)
else:
    print("result",result)
finally:
    print("This will run in any case:")
```
