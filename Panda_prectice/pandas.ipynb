```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Nitesh', 'Rohan', 'Amit', 'Priya', 'Simran', 'Kunal'],
    'Age': [20, 22, 21, 23, 22, 24],
    'Gender': ['M', 'M', 'M', 'F', 'F', 'M'],
    'Maths': [85, 90, 88, 92, 70, 60],
    'Science': [80, 89, 85, 95, 68, 75],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Delhi', 'Mumbai'],
    'Marks': [85, 90, 88, 93, 72, 68],
    'Date_of_Joining': ['2020-01-10', '2019-05-12', '2021-03-09', '2018-12-01', '2020-08-15', '2019-11-20']
})

df
```

```python
df['Total']=df['Maths']+df['Science']
df['doubleMarks']=df['Marks']*2
df['pass']=df['Marks']>40
df
```

```python
import numpy as np

conditions = [
    df['Marks'] >= 90,
    df['Marks'] >= 75,
    df['Marks'] >= 50,
    df['Marks'] < 50
]

values = ['A', 'B', 'C', 'F']

df['Grade'] = np.select(conditions, values)
```

```python
gender_map={'M':"Male",'F':'Female'}
df['GenderFull']=df['Gender'].map(gender_map)
df
```

```python
#.assign() → Multiple new columns in one line
df=df.assign(
    Total = df['Maths'] + df['Science'],
    Avg = lambda x: x['Total'] / 2
)
```

```python
df
```

```python
df=df.astype(
    {
        'Age': 'int32',
    'Marks': 'float32',
    'Gender': 'category'
    }
)
```

```python
df['City'] = df['City'].astype('category')
```

```python
df['Age']=pd.to_numeric(df['Age'],downcast='integer')
```

```python
df['Maths']=pd.to_numeric(df['Maths'],downcast='float')
df
```

```python
df = df.drop(columns=['Unnamed: 0', 'level_0'])
```

```python
df.groupby('City').agg({
    'Marks':'sum',
    'Maths':'mean',
    'Science':'mean'
})
```
