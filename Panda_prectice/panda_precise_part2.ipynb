```python
import pandas as pd

df = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E", "F"],
    "age": [12, 19, 25, 45, 67, 33],
    "gender": ["Male", "Female", "Female", "Male", "Female", "Male"],
    "city": ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai", "Delhi"],
    "math": [78, 90, 55, 89, 45, 70],
    "science": [88, 92, 60, 94, 50, 80],
    "english": [75, 85, 65, 93, 52, 78],
    "salary": [15000, 18000, 45000, 90000, 76000, 65000],
    "height": [1.50, 1.65, 1.70, 1.72, 1.60, 1.80],
    "weight": [45, 60, 70, 82, 55, 78],
    "join_date": ["2022-05-01", "2021/06/12", "13-07-2020", "hello", "2020-01-01", "2019/03/02"]
})
df
```

```python
# ⭐ A. Creating New Columns (Arithmetic + Boolean)
# 1️⃣ Total marks = math + science + english
df["Total_makrs"]=df["math"]+df["science"]+df["english"]
print(df)
```

```python
# 2️⃣ Percentage = total / 3
df["Percentage"]=df["Total_makrs"]/3
print(df["Percentage"])
print(df)
```

```python
# 3️⃣ Result column: marks >= 40 → "Pass" else "Fail"
print(df)
df["is_pass"] = df["Total_makrs"] > 55
df["result"]=df["is_pass"].map({True:"Pass",False:"Fail"})
print(df["result"])
```

```python
# 4️⃣ Adult column: age >= 18 (True/False)
df["is_age"]=df["age"]>18
print(df)
```

```python
# 5️⃣ Discount price = price – (price * 0.10)
df["Discount_Price"]=df["salary"]-(df["salary"]*0.10)
df["10%"]=df["salary"]*0.10
print(df)
```

```python
# 6️⃣ BMI = weight / (height**2)
df["BMI"]=df["weight"]/df["height"]**2
print(df)
```

```python
# 7️⃣ Salary hike = salary * 1.15
Salary_hike=df["salary"]*1.15
print(Salary_hike)
```

```python
# 8️⃣ High_income = salary > 80000
High_income=df["salary"]>80000
print(High_income)
```

```python
# 9️⃣ expensive = price > average price (compute mean first)
means=df["salary"].mean()
df["expensive"]=df['salary']>means
print(df)
```

```python
# ⭐ B. np.select() – Multi Conditions

# 11️⃣ Grade column:
# = 90 ⇒ A
# = 75 ⇒ B
# = 50 ⇒ C
# else Fail
import numpy as np
condition=[
    df["math"]>=90,
    df["math"]>=75,
    df["math"]>=50
]
choice=["A","B","c"]
df["Grade"]=np.select(condition,choice,default="Fail")
print(df)
```

```python
# 12️⃣ Age group:

# < 13 ⇒ Child
# < 20 ⇒ Teenager
# < 60 ⇒ Adult
# else ⇒ Senior
condition=[
    df["age"]<13,
    df["age"]<20,
    df["age"]<60
] 
choice=["Child","Teenager","Adult"]
df["Age_Group"]=np.select(condition,choice,default="Senior")
print(df)
```

```python

# 13️⃣ Salary level:
# < 20000 ⇒ Low
# < 60000 ⇒ Medium
# else ⇒ High
condition=[
    df["salary"]<20000,
    df["salary"]<60000
]
choice=["Low","Medium"]
df["salary_level"]=np.select(condition,choice,default="High")
df
```

```python
# ⭐ C. map() – Category Mapping

# 14️⃣ city_map = {“Delhi”: 1, “Mumbai”: 2, “Pune”: 3} → map() use karke new column banao.
city_map={"Delhi":1,"Mumbai":2,"Pune":3}
df["City_category"]=df["city"].map(city_map)
df
```

```python
# 15️⃣ gender_map = {"Male": 0, "Female": 1}
gender_map={"Male" :0,"Female":1}
df["gender_map"]=df["gender"].map(gender_map)
df
```

```python
# 16️⃣Drop Columns
df.drop(columns=["Percentage"], inplace=True)
df
```

```python
# ⭐ D. assign() – Multiple Columns Together
# 17️⃣ total & percentage ek saath assign() se banao
df.assign(
    total=df["math"]+df["science"]+df["english"],
    percentage= lambda x: x["total"] / 3
)
```

```python
# 18️⃣ is_pass & is_adult assign() se banao
#df.drop(columns=["is_pass"])
df.assign(
    is_pass=df["math"] > 40,
    result = lambda x:x["is_pass"].map({True:"pass",False:"Fail"})
)
df
```

```python
# 19️⃣ 3 columns ek line me banao:
# - after_tax = salary * 0.9
# - doubled_salary = salary * 2
# - yearly_salary = salary * 12
df.assign(
    after_tax=df["salary"]*0.9,
    doubled_salary=df["salary"]*2,
    yearly_salary = df["salary"]*12,
)
```

```python
# ⭐ E. Dtype Conversions + Optimization
# 20️⃣ age column ko int me convert karo
df["age"]=df["age"].astype(int)
print(df["age"])
```

```python
# 21️⃣ marks ko float me convert karo
df["math"]=df["math"].astype(int)
```

```python
# 22️⃣ price ko numeric me convert karo errors='coerce'
df["salary"]=pd.to_numeric(df["salary"],errors="coerce")
df
```

```python
# 23️⃣ date ko datetime me convert karo
df["join_date"]=pd.to_datetime(df["join_date"],errors="coerce")
```

```python
# 24️⃣ city ko category dtype me convert karo
df["city"] = df["city"].astype("category")
```

```python
# 25️⃣ salary column ko downcast integer me convert karo
df["salary"]=pd.to_numeric(df["salary"],downcast="integer")
```
