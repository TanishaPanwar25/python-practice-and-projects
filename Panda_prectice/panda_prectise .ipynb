```python
#chain where 
#50 se chhote values → inme se bhi sirf odd numbers rakho, baaki 0
s=pd.Series([15,25,37,40,50,60,70,80,90,25,35,65])

i=s.where(s<50,0).where(s%2!=0,0)
print(i)
```

```python
# 3 se divisible numbers ko rakho → fir even numbers hi rakho
s=pd.Series([5,10,15,20,25,30,35,40,45,50])
t=s.where(s%3==0,0).where(s%2==0,0)
print(t)
```

```python
#Positive numbers rakho, baaki -1 karo → fir >10 wale hi rakho
s=pd.Series([10,20,-5,-62,-7,55])
ss=s.where(s>0,-1).where(s>10)
```

```python
import pandas as pd
df = pd.DataFrame({
    "name": ["A", "B", "C", "A", "D", None, "E", "F", "F"],
    "age": ["20", None, "22", "20", "-5", "30", "NaN", "25", "25"],
    "marks": [50, 60, None, -10, 70, 85, 85, "90", "90"],
    "city": ["Delhi", "Mumbai", None, "Delhi", "Pune", "Delhi", "Pune", "123", "Pune"],
    "date": ["2024-01-02", "02-03-2024", None, "2024/01/02", "5-13-2024", "2024-13-02", "2024-01-02", "2024-01-02", "2024-01-02"]
})
print(df)
```

```python
# A. Missing Data (isna, dropna, fillna, replace)
# Ek DataFrame me kaunse columns me NaN hain, count nikal.
s=df.isna().sum()
print(s)
# Rows me jisme koi NaN ho sirf wo dikhao.
s=df.isna().any(axis=1)
print(s)
# Columns me jo NaN hain unko mean ya median se fill karo (numeric).
df = df.fillna(df.mean(numeric_only=True))
print(df)
```

```python
# String column ke NaN ko "Missing" se replace karo.
string_col=df.select_dtypes(include=object)
df[string_col.columns]=string_col.fillna("Missing")
print(df)
```

```python
import pandas as pd
df = pd.DataFrame({
    "name": ["A", "B", "C", "A", "D", None, "E", "F", "F"],
    "age": ["20", None, "22", "20", "-5", "30", "NaN", "25", "25"],
    "marks": [50, 60, None, -10, 70, 85, 85, "90", "90"],
    "city": ["Delhi", "Mumbai", None, "Delhi", "Pune", "Delhi", "Pune", "123", "Pune"],
    "date": ["2024-01-02", "02-03-2024", None, "2024/01/02", "5-13-2024", "2024-13-02", "2024-01-02", "2024-01-02", "2024-01-02"]
})
print(df)
```

```python
# C. Removing Duplicates (duplicated, drop_duplicates)
# Duplicate rows check karo.
d=df.duplicated()
print(d)
# Duplicate rows remove karo, keep='first'.
d=df.drop_duplicates(keep="first")
# Duplicate rows remove karo, keep='last'.
d=df.drop_duplicates(keep="last")
# Index reset karo after removing duplicates.
d=df.drop_duplicates(ignore_index=True)
# Duplicate remove karo subset column ke basis par (subset=['name','city']).
d = df.drop_duplicates(subset=["name", "city"], keep="first")
print(d)
```
