```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "days": ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"],
    "sales": [120, 150, 170, 160, 180, 200, 220],
    "customers": [30, 35, 40, 38, 45, 60, 70],
    "profit": [20, 30, 35, 32, 40, 50, 55]
})

df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,4))
plt.title("Daily sale treands")
plt.plot(df["days"],df["sales"],marker="o")
plt.xlabel("days")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,4))
plt.bar(df["days"],df["customers"],color="green")
plt.grid(True)
plt.title("Bar chart")
plt.xlabel("Day")
plt.ylabel("Customers")
plt.show()
```

```python
#BARH — Same data (Horizontal)
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
plt.barh(df["days"],df["customers"],color="purple")
plt.title("Barh Horizontal")
plt.xlabel("customers")
plt.ylabel("Days")
plt.show()
```

```python
#HISTOGRAM — Sales Distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(3,4))
plt.hist(df["sales"], color="skyblue",bins=15,edgecolor="black")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.boxplot(df["sales"])
plt.title("Profit Boxplot")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.scatter(df["customers"], df["sales"], color="red")
plt.title("Customers vs Sales")
plt.xlabel("Customers")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.plot(df["days"],df["sales"],color="green",marker="o",linestyle="--")
plt.title("Customized Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python

fig,ax=plt.subplots(2,2,figsize=(8,6))
# Line
ax[0,0].plot(df["days"],df["sales"],color="skyblue",marker="o")
ax[0,0].set_title("Line chart")
# Bar
ax[0,1].bar(df["days"],df["sales"],color="green",edgecolor="red")
ax[0,1].set_title("Bar chart")
# Histogram
ax[1,0].hist(df["sales"],color="yellow",bins=5,edgecolor="black")
ax[1,0].set_title("Histogram chart")
# Scatter
ax[1,1].scatter(df["sales"],df["customers"],color="pink")
ax[1,1].set_title("Scatter chart")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(10,4))

# Seaborn plot
sns.barplot(x="days", y="sales", data=df, ax=ax[0])
ax[0].set_title("Seaborn Barplot")

# Matplotlib plot
ax[1].bar(df["days"], df["sales"], color="green",edgecolor="yellow")
ax[1].set_title("Matplotlib Bar")

plt.tight_layout()
plt.show()
```
