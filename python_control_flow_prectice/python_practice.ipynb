## Write A Python Program To Get Different Information Of Students From User And Display 
All The Student Information In This Order:
Name ---------
Father name ---------
CNIC ---------
Age ---------
Contact ---------

```python
class Student:
    def __init__(self,name,father_name,cnic,age,contact):
        self.name=name
        self.father_name=father_name
        self.cnic=cnic
        self.age=age
        self.contact=contact
    def display(self):
        print(f"Name ________{self.name}")
        print(f"Father Name -{self.father_name}")
        print(f"cnic ________{self.cnic}")
        print(f"Age __________{self.age}")
        print(f"Contact________{self.contact}")
def main():
    students =[]
    num_student =int(input("Enter the number of students"))

    for i in range(num_student):
        print(f"\n enter details for student {i + 1}:")
        name=input("Enter name")
        father_name=input("Enter Father name")
        cnic=input("Enter Cnic")
        age =input("Enter age")
        contact = input("Enter contact:")
        students.append(Student(name,father_name,cnic,age,contact))
    print("\n student information")
    for student in students:
        student.display()

if __name__ == "__main__":
    main()
```

### Write A Python Program To Count Prime Numbers From 1 To 100

```python
prime_count=0
not_prime_count=0
for i in range(1,101):
    if i==0 or i==1:
        print("Not Prime")
        not_prime_count+=1
    else:
        for j in range(2,i):
            if i%j == 0:
                print(f"{i} Not prime")
                not_prime_count+=1
                break
        else:
            print(f"{i} primee")
            prime_count+=1
print(f"Total prime number between:1 To 100 :- {prime_count} \n Total not prime number :{not_prime_count} ")
```
