#PART 3:  FUNCTION
- 8. Calculator Function – Function calculate(a, b, operation) to do add, sub, mul, div.
- 9. Even or Odd Function – Function that separates even and odd numbers from a list.
- 10. Prime Number Checker – Function that returns True if prime else False.

```python
#Calculator Function – Function calculate(a, b, operation) to do add, sub, mul, div.
def calculate(a,b,operation):
    if operation =='+':
        print("Add", a+b)
    elif operation == "-":
        print("Sub",a-b)
    elif operation == "*":
        print("Mul",a*b)
    elif operation == "/":
        print("Div",a/b)
    else:
        print("Invalid operation")

a=int(input("Enter a number"))
b=int(input("enter second number"))
operation=input("enter operation")
calculate(a,b,operation)
```

```python
#Prime Number Checker – Function that returns True if prime else False.
n=int(input("Enter a number"))
def Prime_Checker(n):
    if n<2:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False
        return True
        
Prime_Checker(n)
        
```

```python
##### 11. ATM Simulator – Balance starts at  ■5000, options to withdraw, check balance, exit.

def ATM():
    print("ATM Simulator")
    balance=5000
    while True:
        print("case 1 for show at start blance \n case 2 for option to withdraw \n case 3 for check balance  \n case 4 for exit")
        Number=int(input("Enter number to perform case (1-4)"))
        
        match Number:
            case 1:
                print(f"start Balance at {balance}")
            case 2:
                print("===========Withdraw==========")
                amount=int(input("Enter amount which you want to withdraw"))
                balance -= amount
                print(f" withdraw successful ! Remaining balance {balance}")
            case 3:
                print("=========Check Balance============")
                print(f"Your current balance {balance}")
            case 4:
                print("Thank You for using ATM .Good Bye/n Exit")
                break;
                
ATM()

        
```
