## ******** Friendship Calculator ********
💪 This is a fun challenge! 💪
You are going to write a function called calculate_friendship_score() that tests the friendship 
compatibility between two people.
1. To work out the friendship score between two names:
2. Take both people's names and check for the number of times the letters in the word FRIEND 
occur.
3. Then check for the number of times the letters in the word BOND occur.
4. Combine these two totals to make a 2-digit number, which represents the Friendship Score.

```python
#First
def calculate_friendship_score(F1,F2):
    word1_count,word2_count=0,0
    combine_word=F1+F2
    for i in 'FRIEND':word1_count+=combine_word.count(i)
    for i in 'BOND':word2_count+=combine_word.count(i)
    return (str(word1_count)+str(word2_count))
Friend1_Name=input("Enter first Friend Name whom you want to see friendship score :")
Friend2_Name=input("Enter Second Friend Name whom you want to see friendship score: ")
F1=Friend1_Name.upper()
F2=Friend2_Name.upper()
print("*******************")
print("Freindship Score is :" + calculate_friendship_score(F1,F2))
```

### ******** Guess the Number Game  ********
 Objective
The goal of the game is to guess a randomly generated number within a specific range. The 
program gives hints whether your guess is too high or too low until you find the correct 
number.
�How the Game Works
1. The program picks a random number between a given range (for example, 1 to 100).
2. The player tries to guess the number by typing input.
3. The program checks the player's guess and provides feedback:
 • 'Too High' → if the guess is greater than the number.
 • 'Too Low' → if the guess is smaller than the number.
4. This continues until the player guesses correctly.
5. The game shows the total number of attempts it took to guess right.

```python
import random
i=random.randint(1,100)
print("Random Number:",i)
count=1
while True:
    guess_number=int(input("Guess any number"))
    if i != guess_number:
        if guess_number>i:
            print("Too High")
        else:
            print("Too Low")
        count+=1
        continue
    else:
        print("congratulation! your guess correct")
        break

print(f"you taken {count} attempt")
```

```python
# Game
import random

print("******** BLACKJACK GAME ********")
card = []
p1 = []
d1 = []
Current_Score_Player = 0
Current_Score_Dealer = 0

print("Card Type:- Hearts(♥), Diamonds(♦), Clubs(♣), Spades(♠)")

A_card = ['J', 'Q', 'K', 'A']
card_type = ['♥', '♦', '♣', '♠']

for ct in card_type:
    for i in range(2, 11):
        card.append(f"{i}{ct}")
    for j in A_card:
        card.append(f"{j}{ct}")

def Number_convertor(c1, Current_Player_Score, who):
    if c1[0] in ['J', 'Q', 'K']:
        Current_Player_Score += 10
    elif c1[0] == 'A':
        Current_Player_Score += 11
    else:
        Current_Player_Score += int(c1[:-1])
    

    if Current_Player_Score > 21:
        num_aces = sum(1 for card_ in (p1 if who == "Player" else d1) if card_[0] == 'A')
        while Current_Player_Score > 21 and num_aces > 0:
            Current_Player_Score -= 10
            num_aces -= 1
    
    print(f"{who} Current Score:", Current_Player_Score)
    return Current_Player_Score

while len(p1) < 2:
    p_card = random.choice(card)
    if p_card not in p1:
        p1.append(p_card)
        Current_Score_Player = Number_convertor(p_card, Current_Score_Player, who="Player")
        if Current_Score_Player == 21:
            print(" BLACKJACK !")
            break

while len(d1) < 2:
    d_card = random.choice(card)
    if d_card not in d1:
        d1.append(d_card)
        Current_Score_Dealer = Number_convertor(d_card, Current_Score_Dealer, who="Dealer")
        if Current_Score_Dealer == 21:
            print(" BLACKJACK !")
            break

print("Player cards:", p1)
print("Dealer cards:", d1)

def Hitstand(Current_Score_Player):
    while True:
        if Current_Score_Dealer == 21:
            break
        player_decision = input("Enter Hit to draw a card or Stand to stop: ").strip().lower()
        if player_decision == 'hit':
            while True:
                c1 = random.choice(card)
                if c1 not in p1 and c1 not in d1:
                    break
            p1.append(c1)
            print("Player card List:", p1)
            Current_Score_Player = Number_convertor(c1, Current_Score_Player, "Player")
            if Current_Score_Player > 21:
                print("Your score is greater than 21 / Bust!")
                break
        elif player_decision == 'stand':
            break
        else:
            print("Invalid input! Enter Hit or Stand")
    return Current_Score_Player

Current_Score_Player = Hitstand(Current_Score_Player)

print("******* Dealer ********")

while Current_Score_Dealer < 18:
    c1 = random.choice(card)
    if c1 in d1 or c1 in p1:
        continue
    d1.append(c1)
    print("Dealer card List:", d1)
    Current_Score_Dealer = Number_convertor(c1, Current_Score_Dealer, "Dealer")
    if Current_Score_Dealer > 21:
        print("Dealer Busted!")

print("Final Scores:")
print("Player:", Current_Score_Player)
print("Dealer:", Current_Score_Dealer)

if Current_Score_Player > 21 and Current_Score_Dealer > 21:
    print("Busted Game!")
elif Current_Score_Player > 21:
    print("Player Busted! Dealer Wins!")
elif Current_Score_Dealer > 21:
    print("Dealer Busted! Player Wins!")
elif Current_Score_Player > Current_Score_Dealer:
    print("Player Wins!")
elif Current_Score_Player < Current_Score_Dealer:
    print("Dealer Wins!")
else:
    print("It's a Tie!")
```
