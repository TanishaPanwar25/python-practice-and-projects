# ****Higher and lower Game**** : 
    use the below data and ask user among these to celebrities who has the highest followers
    if correct add +1 to the score and else 0 and keep asking till 
    the user wants to play and finally show the score( or you can keep showing the score as you play) 

```python
import random

with open("Data.txt", "r") as f:
    data = eval(f.read()) 

pr=[]
score=0
print("***************Higher and lower Game**************")
time=1
while True:
    celebr1=random.choice(data)
    celebr2=random.choice(data)
    print("==================Celebraties 1  =================")
    for key in celebr1:
        if key=='follower_count':
            continue
        value=celebr1[key]
        print(f"{key}:{value}")

    print("==================Celebraties 2  =================")
    for key in celebr2:
        if key=='follower_count':
            continue
        value=celebr2[key]
        print(f"{key}:{value}")
    
    user_input=input("Enter A for Celebraties 1 and B for Celebraties 2 and enter any key for exit")
    if user_input=='A':
        if celebr1['follower_count']>celebr2['follower_count']:
            print("celebr1 Higher Follower")
            score+=1
        else:
            print("wrong guess")
            score+=0
    elif user_input == 'B':
        if celebr1['follower_count'] < celebr2['follower_count']:
            print("celebr2 Higher Follower")
            score+=1
        else:
            print("wrong guess")
            score+=0
    else:
        print("Game Over")
        print(f"Your total score: {score} ,{time} time's you play")
        break
    print("Score",score)
    time+=1
```
